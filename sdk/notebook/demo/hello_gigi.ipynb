# Hello, GIGI

A first run through the **gigi** Jupyter kernel. Make sure the kernel is selected from the menu (“GIGI (GQL)”).

Before running any cells:

- Set `GIGI_URL` to your engine (default: `http://localhost:3142`).
- Set `GIGI_API_KEY` to your X-API-Key value.

You can override either inline with `%env`.

```gql
%env GIGI_URL=https://gigi-stream.fly.dev
```

```gql
%%config
```

## The default cell language is GQL

Just type GIGI Query Language and run. Output is the engine’s parsed JSON, rendered as a Python dict / object.

```gql
SHOW BUNDLES
```

## Causal-States commutator — the new endpoint

Use `%%commutator` to POST a JSON body to `/v1/causal_states/commutator`. The result renders as a regime-tagged table:

- `smooth` — KL is finite, TV is small. The noisy HMM lives here.
- `sofic` — KL is divergent. The Even Process lives here.
- `borderline` — TV in the crossover band.

```gql
%%commutator
{
  "a": {"kind": "hmm", "alpha": 0.2, "beta": 0.3, "symbol": 0},
  "b": {"kind": "hmm", "alpha": 0.2, "beta": 0.3, "symbol": 1},
  "base_belief": [0.5, 0.5]
}
```

```gql
%%commutator
{
  "a": {"kind": "even_u0"},
  "b": {"kind": "even_u1"},
  "base_belief": [0.6666666666666666, 0.3333333333333333]
}
```
